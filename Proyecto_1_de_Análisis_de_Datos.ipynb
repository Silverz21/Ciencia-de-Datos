1.  Hacer una gráfica del promedio de Diaphorinas por trampa para cada mes de  los  años  2015, 2017, 2020. Es  decir  una  gráfica  por  cada  año  donde el eje horizontal esté el mes y el vertical el promedio de diaphorinas por trampa.

```python
path = 'capturas_2015-2020.xlsx'
df = pd.read_excel(path)
df=df.set_index('Mes')
prom2015 = pd.DataFrame(df.groupby(level=['Mes'])['No_DiaphorinasXtrampa'].mean())
prom2015.rename(index={1.0:'Enero', 2.0:'Febrero', 3.0:'Marzo',
                       4.0:'Abril', 5.0:'Mayo', 6.0:'Junio',
                       7.0:'Julio', 8.0:'Agosto', 9.0:'Septiembre', 
                       10.0:'Octubre', 11.0:'Noviembre', 12.0:'Diciembre'},
                columns={"No_DiaphorinasXtrampa":''},
                inplace=True) 
```

```python
graf2015 = prom2015.plot(ylabel="Promedio x Trampa", xlabel="Mes", kind="bar", 
              title="Promedio de Diaphorinas por trampa por cada mes",
              )
```

3. Hacer una tabla donde la primera columna sea el mes, la segunda columna sea 
el índice Id (por trampa) y la tercera el índice Cx para el año 2015.

```python
prom2015
```

```python
var2015 = pd.DataFrame(df.groupby(level=['Mes'])['No_DiaphorinasXtrampa'].var())
var2015.rename(index={1.0:'Enero', 2.0:'Febrero', 3.0:'Marzo',
                       4.0:'Abril', 5.0:'Mayo', 6.0:'Junio',
                       7.0:'Julio', 8.0:'Agosto', 9.0:'Septiembre', 
                       10.0:'Octubre', 11.0:'Noviembre', 12.0:'Diciembre'},
                columns={"No_DiaphorinasXtrampa":''},
                inplace=True) 
var2015
```

```python
nueva_tabla = prom2015/var2015
nueva_tabla.rename( columns={"":'Indice de Dispersión (Id)'},
                inplace=True) 
nueva_tabla
```

```python
sum2015 = pd.DataFrame(df.groupby(level=['Mes'])['No_DiaphorinasXtrampa'].sum())
sum2015.rename(index={1.0:'Enero', 2.0:'Febrero', 3.0:'Marzo',
                       4.0:'Abril', 5.0:'Mayo', 6.0:'Junio',
                       7.0:'Julio', 8.0:'Agosto', 9.0:'Septiembre', 
                       10.0:'Octubre', 11.0:'Noviembre', 12.0:'Diciembre'},
                columns={"No_DiaphorinasXtrampa":''},
                inplace=True) 
sum2015
```

```python
nueva_tabla["Coeficiente de Green (C_x)"]= var2015/prom2015 - 1/sum2015-1
nueva_tabla
```
